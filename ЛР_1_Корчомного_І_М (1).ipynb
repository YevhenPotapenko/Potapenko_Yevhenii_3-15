```python
matrix1 = np.random.randint(1,11,size=(4,4))
matrix2 = np.random.randint(1,11,size=(4,4))

print("Матриця 1:")
print(matrix1)

print("\nМатриця 2:")
print(matrix2)
```

```python
matrix_sum = matrix1 + matrix2

matrix_diff = matrix1 - matrix2

matrix_product = matrix1 * matrix2

matrix_division = np.divide(matrix1, matrix2, where=(matrix2!=0))

print("\nmatrix_sum")
print(matrix_sum)

print("\nmatrix_diff")
print(matrix_diff)

print("\nmatrix_product")
print(matrix_product)

print("\nmatrix_division")
print(matrix_division)
```

```python
a = np.array([[1,3,-4],[0,2,-3],[-2,2,5]])
b = np.array([5,3,4])

x= np.linalg.solve(a,b)
print('Вбудована функція', x.round(1))

inv_a = np.linalg.inv(a)
x1 = inv_a.dot(b)
print('Матричний метод', x1.round(1))
```
